```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
```

```python
# Load the CSV file
data = pd.read_csv("CEC_urban_data.csv")

y = data['GD Rate'] = 100* data['41_votes'] / data['total_votes']
x = data['Turnout Rate'] = 100*  data['total_votes'] / data['total_registered_voters']
#data['GD Turnout'] = data['41_votes'] / data['total_registered_voters']

# remove invalid data points (for example, data points where 'total_registered_voters' is 0)
mask = ~np.isnan(x) & ~np.isnan(y) & ~np.isinf(x) & ~np.isinf(y)
x = x[mask]
y = y[mask]

# Create a scatter plot
plt.figure(figsize=(10, 10)) 
plt.scatter(x, y, s=15, alpha=0.35, color='maroon')
plt.xlabel('Turnout (%)')
plt.ylabel('Georgian Dream Vote (%)')
plt.title('Georgian Election 2024: Urban Polling Stations (N = {})'.format(len(x)))


plt.gca().set_xticks(np.arange(0, 1.001*100, 0.05*100))
plt.gca().set_yticks(np.arange(0, 1.001*100, 0.05*100))
plt.grid(True, linestyle='dashed')

plt.xlim(0, 100)  # Adjust to your desired x-range
plt.ylim(0, 100)  # Adjust to your desired y-range


plt.show()
```

```python
# Load the CSV file
data = pd.read_csv("CEC_rural_data.csv")

y = data['GD Rate'] = 100* data['41_votes'] / data['total_votes']
x = data['Turnout Rate'] = 100*  data['total_votes'] / data['total_registered_voters']
#data['GD Turnout'] = data['41_votes'] / data['total_registered_voters']

# remove invalid data points (for example, data points where 'total_registered_voters' is 0)
mask = ~np.isnan(x) & ~np.isnan(y) & ~np.isinf(x) & ~np.isinf(y)
x = x[mask]
y = y[mask]

# Define the grid size
grid_size = 100

# Create a 2D histogram of the data with 100x100 bins
hist, xedges, yedges = np.histogram2d(data['Turnout Rate'], data['GD Rate'], bins=grid_size, range=[[0, 100], [0, 100]])

# Plot the heatmap
plt.figure(figsize=(15, 15))
plt.imshow(hist.T, origin='lower', cmap='plasma', extent=[0, 100, 0, 100])

# Add colorbar to indicate frequency
plt.colorbar(label='Frequency', fraction=0.046, pad=0.04)

# Add labels and title
plt.xlabel('Turnout (%)')
plt.ylabel('Georgian Dream Vote (%)')
plt.title('Georgian Election 2024: Rural Polling Stations (N = {})'.format(len(x)))

plt.show()
```
